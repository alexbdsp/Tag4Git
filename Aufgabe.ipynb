Aufgabe: Erstellen Sie ein Streudiagramm, das die Beziehung zwischen Temperatur und Luftfeuchtigkeit f√ºr das Jahr 2024 darstellt.  Verwenden Sie Pandas und Matplotlib, um das Diagramm zu erstellen.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Zuf√§llige Daten f√ºr Temperatur und Luftfeuchtigkeit
np.random.seed(0)
tage = pd.date_range(start='2024-01-01', end='2024-12-31')
temperatur = np.random.normal(loc=20, scale=5, size=len(tage))
luftfeuchtigkeit = np.random.normal(loc=50, scale=10, size=len(tage))

# DataFrame mit den generierten Daten
daten = pd.DataFrame({'Datum': tage, 'Temperatur (C)': temperatur, 'Luftfeuchtigkeit (%)': luftfeuchtigkeit})

# Streudiagramm erstellen
# Dein Code:
plt.figure(figsize=(10, 6))
plt.scatter(daten['Temperatur (C)'], daten['Luftfeuchtigkeit (%)'], color='blue', alpha=0.5)
plt.title('Beziehung zwischen Temperatur und Luftfeuchtigkeit')
plt.xlabel('Temperatur (C)')
plt.ylabel('Luftfeuchtigkeit (%)')
plt.grid(True)
plt.show()
```
